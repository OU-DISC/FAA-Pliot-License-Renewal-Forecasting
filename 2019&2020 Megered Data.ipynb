```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Load two dataset together

```python
d = pd.read_csv('2019_data.csv')
d.drop(columns =['Unnamed: 0'], inplace=True)
d.head()
```

```python
d1 = pd.read_csv('2020_data.csv')
d1.drop(columns =['Unnamed: 0'], inplace=True)
d1.head()
```

## Concate two dataset together

```python
df = pd.concat([d, d1])
```

```python
df.head(20)
```

```python
df.shape
```

## Unique values of the data

```python
df.nunique()
```

## Drop duplicates

```python
df.drop_duplicates(keep='first', inplace=True)
```

## Date types

```python
df.info()
```

### Convert float to integer

```python
df[['ENROLID', 'DOBYR', 'YEAR', 'AGE', 'SEX', 'NDCNUM']] = df[['ENROLID', 'DOBYR', 'YEAR', 'AGE', 'SEX', 'NDCNUM']].astype(int)
```

```python
df.info()
```

### Missing Values

```python
df.isnull().sum()
```

### Visualize the Null Value

```python
plt.figure(figsize = (10, 7))
sns.heatmap(df.isnull(), yticklabels=False, cbar = False, cmap='viridis')
```

### Fill Null value with '999'

```python
df = df.fillna('999')
```

### Visualize the null value again

```python
plt.figure(figsize = (10, 7))
sns.heatmap(df.isnull(), yticklabels=False, cbar = False, cmap='viridis')
```

## Plots

### Countplot of diagnosis codes relates to heart disease

```python
dx = df.loc[df['DX1'].str.startswith('I')]
dx['DX1'].value_counts()
```

### DX1 Heart Disease Diagnosis

```python
# DX1 
plt.figure(figsize=(14, 7))

sns.countplot(x=dx['DX1'][:300], hue='SEX', edgecolor=".7", palette='RdBu_r', data=df)
plt.tick_params(labelrotation=90)
plt.legend(loc='upper right', title='SEX')
```

### DX2 Heart Disease Diagnosis

```python
# DX2
dx = df.loc[df['DX2'].str.startswith('I')]
plt.figure(figsize=(14, 7))
sns.countplot(x=dx['DX2'][:200], hue='SEX', edgecolor=".7", palette='RdBu_r', data=df)
plt.xticks(rotation=90)
plt.legend(loc='upper right', title='SEX')
```

### All kind of Disease Diagnosis code 

```python
col = [x for x in df.columns if x.startswith('DX')]
```

```python
fig, ax = plt.subplots(3, 4, figsize=(20, 15))
ax = ax.ravel()
position = 0

for i in col[2:14]:
    order = df[i].value_counts().iloc[:20].index
    sns.countplot(x=i, ax=ax[position], hue='SEX', palette='RdBu', order=order, data=df)
    ax[position].tick_params(labelrotation=90)

    ax[position].set_title(i, fontdict={'fontsize': 20})
    position += 1
    
plt.subplots_adjust(hspace=0.9)
plt.show()
```
