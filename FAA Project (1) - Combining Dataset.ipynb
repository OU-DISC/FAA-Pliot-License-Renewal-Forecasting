## Combine Inpatient Admission Dataset (2019 & 2020)

### Load in 2019 Inpatient Admission Data

```python
columns = ['ENROLID','YEAR', 'AGE', 'DX1', 'DX2', 
           'DX3', 'DX4', 'DX5', 'DX6', 'DX7', 
           'DX8', 'DX9', 'DX10', 'DX11', 'DX12', 
           'DX13', 'DX14', 'DX15','SEX']

# skiprows = 3855, before that all Enrolid is NaN
df = pd.read_csv('ccaei192_a.csv', skiprows = range(1, 3855),
                                       usecols = columns)

df.head()
```

### Load in 2020 Inpatient Admission Dataset

```python
df1= pd.read_csv('ccaei200_a.csv', skiprows = range(1, 3458),
                                    usecols = columns)

df.head()
```

### Merge Two Inpatient Admission Dataset Together

```python
dff = pd.merge(df, df1, on=['ENROLID', 'SEX'])
```

```python
# Drop the duplicates
dff = dff.drop_duplicates(keep='first')
```

```python
dff.head()
```

```python
dff.shape
```

## Combine Perscription Drug Datasets

### Load in 2019 Perscription Drug Data

```python
col = ['ENROLID', 'NDCNUM']

df_d19 = pd.read_csv('ccaed192_a.csv', skiprows = range(1, 1000000),
                                       usecols = col)

df_d19.head()
```

### Load in 2020 Perscription Drug Data

```python
df_d20 = pd.read_csv('ccaed200_a.csv', skiprows = range(1, 1000000),
                                        usecols = col)

df_d20.head()
```

## Concate Two Perscription Drug Data

```python
d = pd.concat([df_d19, df_d20])
```

### Load in red book data

```python
red = pd.read_csv('redbook.csv', usecols = ['NDCNUM', 'PRODNME'])
```

## Merge Perscription Drug Tables and RedBook Table

```python
dd = pd.merge(red,d, on='NDCNUM')
```

```python
dd.shape
```

```python
# Drop Duplicates
dd.drop_duplicates(keep='first', inplace=True)
```
